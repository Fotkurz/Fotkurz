## Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?

A classe tratamento possuí duas variáveis, com_controle e com_tratamento. 
A abordagem com_controle foi feita utilizando-se apenas um composto que foi o 'cacb2b860'.
Isso ocorre porque a abordagem utilizada testou as drogas do tratamento ao menos 6 vezes, cada uma. Se houvesse um teste variando entre controle e tratamento, as drogas teriam que ser testadas mais 6 vezes, levando em consideração que ainda existem as variaveis dose e tempo. Nosso data frame seria ainda muito maior. Quem coletou os dados decidiu não realizar uma abordagem assim. Por isso a variável tratamento ficou bem desbalanceada.

```python
dados_comcontrole = dados.query('tratamento == "com_controle"')
dados_comcontrole
```

```python
dados_comcontrole.query('droga != "cacb2b860"')
```

## Desafio 03: Proporção das classes tratamento.

```python
# Normalizando os valores (achando a distribuição de 0~1 da quantidade de valores)
total = dados['tratamento'].value_counts(normalize=True)
```

```python
total_perc = total*100
print("A porcentagem está distribuída entre:")
print("Com droga: %.2f" % (total_perc[0]), "%")
print("Com tratamento: %.2f" % (total_perc[1]), "%")
```

## Desafio 04: Quantas tipos de drogas foram investigados.

Podemos resolver essa usando o método unique, que deve criar uma série com todos os valores unicos e em seguida usar o size para retornar a quantidade de valores do unique.

## Desafio 06: Renomear as colunas tirando o hífen.

Como há 877 colunas, o método mais fácil seria usando o replace.
Pra isso precisamos no entando usar o método str que é uma interface para diversos métodos relacionados a strings como o próprio replace.
Em seguida basta passar como parâmetro para o replace o caracter que queremos trocar, no caso o '-' por '' (vazio), e em seguida, armazenar em uma variavel nova, eu vou usar o próprio dados.columns para fazer isso no nosso próprio DataFrame.

```python
dados.columns = dados.columns.str.replace('-',"")
```

## Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)

Vou usar o próprio matplotlib.pyplot para tratar o piechart.
Setei a variável explode para separar um pouco a fatia menor da maior e apliquei uma sombra (shadow) só para ficar mais bonito.
Defini como parâmetro as labels das áreas do gráfico, título e também a escrita no eixo y.

```python
pieplot_tratamento = dados['tratamento'].value_counts().plot.pie(explode=(0,0.2),shadow=True, figsize=(10,5), labels=['Com Droga', 'Com Tratamento'])
pieplot_tratamento.set_title("Distribuição do tratamento")
pieplot_tratamento.set_ylabel("Tratamento")
```

No gráfico de barras vou usar o seaborn, um framework especializado em criar gráficos.

```python
# importando o seaborn, é costume usar sns como variavel
import seaborn as sns
data_sns = dados['tempo'].value_counts()
data_sns
```

```python
grafico = sns.barplot(x=data_sns.index, y=data_sns)
grafico.set_title("Tempo de exposição do tratamento")
grafico.set_xlabel("Horas")
grafico.set_ylabel("Quantidade")
```
